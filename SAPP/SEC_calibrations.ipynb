# SEC calibration functions

These function calibrate using the Cytiva HMW calibration kit (for S200) and LMW calibration kit (for S75), calibration proteins are hardcoded to each column.

For the agilent system we provide an additional `_fitvoid` function that allows you to fit out a first aggregate peak (that is not part of the calibration kit). This void peak is often aggregate, but not a protein in the calibration set, so we added another expected peak to correct for this peak.
 
for the agilent system
```
S75_5_150_calibration_agilent(path_dextran, path_calib, flowrate, comment="" )
S75_5_150_calibration_agilent_fitvoid(path_dextran, path_calib, flowrate, comment="" )
S200_5_150_calibration_agilent(path_dextran, path_calib, flowrate, comment="" )
S200_5_150_calibration_agilent_fitvoid(path_dextran, path_calib, flowrate, comment="" )
```
for the Akta system
```
S75_5_150_calibration_fitvoid_akta(chroma, no_dextran, no_calib )
S200_5_150_calibration_fitvoid(chroma, no_dextran, no_calib )
```
